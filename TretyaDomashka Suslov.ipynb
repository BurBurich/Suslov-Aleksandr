```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("russian_passenger_air_service.csv")
```

```python
df.groupby('Airport name').groups
```

```python
Ark = df.groupby('Airport name').get_group('Arkhangelsk (Talagy)')
Ark = Ark[Ark['Year']<2020]
Ark
```

```python
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
Ark['December'] = Ark['Whole year']  - Ark[months[:-1]].sum(axis = 1)
K = np.array(Ark[months])[::-1]
K1 = K.flatten()
KInThousand=K1/1000
Data = pd.DataFrame(KInThousand, columns = ['pas'])
Data
```

```python
Data.plot(figsize=(18,8))
```

Проведем тест Дики-Фулера

Нулевая гипотеза, что стационарности нет

```python
test = sm.tsa.adfuller(Data['pas'])
test
```

На всех трех предложенных уровнях нулевая гипотеза принимается - стационарности нет.

Первые разности:

```python
Datadiff1 = Data.diff(periods=1).dropna()
```

Снова Дики-Фулер

```python
test1 = sm.tsa.adfuller(Datadiff1['pas'])
test1
```

Видим, что стационарность появилась.

Нарисуем первые разности

```python
Datadiff1.plot(figsize=(18,8))
```

```python
fig = plt.figure(figsize=(20,12))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(Datadiff1.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(Datadiff1, lags=40, ax=ax2)
```

Ljung-Box (Q): проверяет что остатки образуют белый шум - это нулевая гипотеза - значит p-value должно быть большим, Jarque-Bera (JB) проверяет остатки на нормальность - p-value должно быть тоже большим (ваш комментарий)

```python
M = sm.tsa.statespace.SARIMAX(Data[:144], order=[1, 1, 0], seasonal_order=[4, 1, 1, 12], trend=None).fit()
print(M.summary())
```

Ljung-Box (Q) и Jarque-Bera (JB) имеют хороший p-value.
Теперь строим прогноз

```python
print(M.forecast(12))
```

```python
print(Data[144:])
```

```python
plt.figure(figsize=(20,10))
plt.plot(Data[108:]['pas'],color='blue',label='жизнь')
plt.plot( M.forecast(12), color='red',label='прогноз' )
plt.legend(loc='best', fontsize = 14)
plt.show()
```
